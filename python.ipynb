```python
def outer():
    lst = []
    def inner(x, lst=lst):
        lst.append(x)
        return lst
    return inner(1), inner(2)

print(outer())
```

```python
x = 0
def f():
    x = 1
    def g():
        exec("x = x + 1", locals=locals())
    g()
    return x
print(f())
```
